```python
# Choosing 'math score' as target variable
target = "Math_Score"

# Input features
features = df.drop(columns=[target]).columns.tolist()

print("Target Variable:", target)
print("\nInput Features:")
print(features)
```

```python
# Convert math score into pass/fail for imbalance check
df['math_pass'] = df['Math_Score'].apply(lambda x: 1 if x >= 50 else 0)

df['math_pass'].value_counts()
```
